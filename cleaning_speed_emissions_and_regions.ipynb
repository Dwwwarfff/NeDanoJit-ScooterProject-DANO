```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# импортируем csv'шку
df = pd.read_csv('kicksharing.csv')
```

```python
# чудеса стандартизации названий городов и областей

def replace(target: str, replacement: list[str]):
  for i in replacement:
    df['lvn_state_nm'] = df['lvn_state_nm'].replace({i: target})

print(len(df['lvn_state_nm'].dropna().unique()))

replace('МОСКВА', ['Г. МОСКВА', 'ГОРОД МОСКВА', 'Г МОСКВА', 'МОСКВА Г', 'МОСККОВСКАЯ'])
replace('САНКТ-ПЕТЕРБУРГ', ['Г. САНКТ-ПЕТЕРБУРГ', 'ГОРОД САНКТ-ПЕТЕРБУРГ', 'САНКТ-ПЕТЕРБУРГ Г', 'Г САНКТ-ПЕТЕРБУРГ'])
replace('КЕМЕРОВСКАЯ ОБЛ', ['КЕМЕРОВСКАЯ ОБЛАСТЬ - КУЗБАСС ОБЛ', 'КЕМЕРОВСКАЯ ОБЛАСТЬ - КУЗБАСС'])
replace('ЧУВАШСКАЯ РЕСПУБЛИКА', ['ЧУВАШИЯ ЧУВАШСКАЯ РЕСПУБЛИКА -', 'ЧУВАШСКАЯ РЕСПУБЛИКА ЧУВАШИЯ', 'ЧУВАШСКАЯ РЕСПУБЛИКА - ЧУВАШИЯ'])
replace('СВЕРДЛОВСКАЯ ОБЛ', ['ОБЛ СВЕРДЛОВСКАЯ', 'ОБЛ. СВЕРДЛОВСКАЯ', 'СВЕРДЛОВСКАЯ ОБЛАСТЬ', 'СВЕРДЛОВСКАЯ'])
replace('ГОМЕЛЬСКАЯ ОБЛ', ['ГОМЕЛЬСКАЯ ОБЛАСТЬ', 'ГОМЕЛЬСКАЯ'])
replace('ТАТАРСТАН РЕСП', ['РУСП ТАТАРСТАН', 'РЕСПУБЛИКА ТАТАРСТАН (ТАТАРСТАН)', 'РЕСП. ТАТАРСТАН', 'РЕСП ТАТАРСТАН', 'ТАТАРСТАН', 'РЕСПУБЛИКА ТАТАРСТАН'])
replace('МОСКОВСКАЯ ОБЛ', ['ОБЛ. МОСКОВСКАЯ', 'МОСККОВСКАЯ', 'МОСКОВСКАЯ ОБЛ.', 'ОБЛ МОСКОВСКАЯ', 'МОСКОВСКАЯ ОБЛАСТЬ', 'МОСКОВСКАЯ'])
replace('ДОНЕЦКАЯ НАРОДНАЯ РЕСП', ['РЕСП ДОНЕЦКАЯ НАРОДНАЯ'])
replace('ХАНТЫ-МАНСИЙСКИЙ АО', ['АО ХАНТЫ-МАНСИЙСКИЙ АВТОНОМНЫЙ ОКРУГ - ЮГРА', 'ХАНТЫ-МАНСИЙСКИЙ АВТОНОМНЫЙ ОКРУГ - ЮГРА', 'ХАНТЫ-МАНСИЙСКИЙ АВТОНОМНЫЙ ОКРУГ - ЮГРА АО'])
replace('НОВОСИБИРСКАЯ ОБЛ', ['НОВОСИБИРСКАЯ ОБЛАСТЬ', 'НОВОСИБИРСКАЯ'])
replace('ЧЕЛЯБИНСКАЯ ОБЛ', ['ЧЕЛЯБИНСКАЯ ОБЛ.', 'ЧЕЛЯБИНСКАЯ ОБЛАСТЬ'])
replace('ЛЕНИНГРАДСКАЯ ОБЛ', ['ЛЕНЕНГРАДСКАЯ ОБЛ', 'ЛЕНИНГРАДСКАЯ ОБЛАСТЬ'])
replace('НИЖЕГОРОДСКАЯ ОБЛ', ['НИЖЕГОРОДСКАЯ'])
replace('ИРКУТСКАЯ ОБЛ', ['ИРКУТСКАЯ ОБЛАСТЬ'])
replace('АСТРАХАНСКАЯ ОБЛ', ['АСТРАХАНСКАЯ ОБЛАСТЬ'])
replace('САХАЛИНСКАЯ ОБЛ', ['САХАЛИНСКАЯ ОБЛАСТЬ'])
replace('УДМУРТСКАЯ РЕСП', ['УДМУРТСКАЯ'])
replace('ХАНТЫ-МАНСИЙСКИЙ АО', ['ХАНТЫ-МАНСИЙСКИЙ АВТОНОМНЫЙ ОКРУГ'])
replace('ВОЛГОГРАДСКАЯ ОБЛ', ['ВОЛГОГРАДСКАЯ ОБЛАСТЬ'])
replace('ЯМАЛО-НЕНЕЦКИЙ АО', ['АО ЯМАЛО-НЕНЕЦКИЙ']) 
replace('ТУЛЬСКАЯ ОБЛ', ['ТУЛЬСКАЯ ОБЛАСТЬ'])
replace('САХА /ЯКУТИЯ/ РЕСП', ['САХА /ЯКУТИЯ/'])

for i in df['lvn_state_nm'].dropna().unique():
    if i.startswith('ОБЛ'):
       replace(' '.join(i.split(' ')[1:]) + ' ОБЛ', [i])
    elif i.startswith('РЕСП'):
       replace(' '.join(i.split(' ')[1:]) + ' РЕСП', [i])
    elif i.startswith('ОБЛАСТЬ'):
       replace(' '.join(i.split(' ')[1:]) + ' ОБЛ', [i])
    elif i.startswith('РЕСПУБЛИКА'):
       replace(' '.join(i.split(' ')[1:]) + ' РЕСП', [i])
    elif i.startswith('КРАЙ'):
       replace(' '.join(i.split(' ')[1:]) + ' КРАЙ', [i])
    elif i.startswith('АОБЛ'):
       replace(' '.join(i.split(' ')[1:]) + ' АОБЛ', [i])

print(len(df['lvn_state_nm'].dropna().unique()))
df['lvn_state_nm'].dropna().unique()
```

```python
# смотрим чтобы не было повторов
regions = df['lvn_state_nm'].dropna().unique().tolist()

dict = {}
for region in regions:
    for s in region.split(' '):
        if s in dict:
            dict[s] += 1
        else:
            dict[s] = 1

for k, v in dict.items():
    if (v > 1):
        print(k, v)

regions

with open('regions.txt', 'w') as f:
    for k in regions:
        f.write(f'{k}\n')
```

```python
# приводим к pandas'овскому datetime

df['book_start_dttm'] = pd.to_datetime(df['book_start_dttm'])
df['book_end_dttm'] = pd.to_datetime(df['book_end_dttm'])
df['local_book_start_dttm'] = pd.to_datetime(df['local_book_start_dttm'])
df['created_dttm'] = pd.to_datetime(df['created_dttm'])
```

```python
# вычисляем время поездки

df['travel_time'] = df.book_end_dttm - df.book_start_dttm
df['travel_time_h'] = (df['travel_time'].dt.total_seconds() / 3600)
```

```python
# вычисляем среднюю скорость по расстоянию и времени

df['average_speed'] = df['distance_km'].round(3) / (df['travel_time'].dt.total_seconds() / 3600).round(3)
```

```python
# считаем что скорость самоката не может быть больше 40 км/ч

df = df[df['average_speed'] < 40]
```
