```python
import pandas as pd
import numpy as np
import seaborn as sns

sns.set_theme(rc={'figure.figsize':(11.7,8.27)})
```

```python
df = pd.read_csv('../kicksharing_drop.csv',index_col=0)
```

```python
df['local_book_start_dttm'] = pd.to_datetime(df['local_book_start_dttm'])
```

```python
df['local_hour_start'] = df['local_book_start_dttm'].dt.hour
```

```python
from scipy import stats
```

```python
grouped = df.groupby('local_hour_start')

results = {}
for column in df.select_dtypes(include=['float64', 'int64']).columns:
    group_a = grouped.get_group(5)[column]
    group_b = grouped.get_group(8)[column]

    group_a = group_a[group_a.notnull() & group_a.apply(lambda x: isinstance(x, (int, float)))]
    group_b = group_b[group_b.notnull() & group_b.apply(lambda x: isinstance(x, (int, float)))]

    f_stat, p_value = stats.f_oneway(group_a, group_b)
    results[column] = {'f-statistic': f_stat, 'p-value': p_value}

results_df = pd.DataFrame(results).T

results_df = results_df.sort_values(by='p-value')

results_df
```

```python
df['created_dttm'] = pd.to_datetime(df['created_dttm'])
df['book_start_dttm'] = pd.to_datetime(df['book_start_dttm'])
df
```

```python
df['seconds_difference'] = df['book_start_dttm'] - df['created_dttm']
```

```python
df['seconds_difference'] = df['seconds_difference'].dt.total_seconds()
```

```python
df
```

```python
grouped = df.groupby('transport_model')

results = {}

for model in grouped.groups:
    group_a = grouped.get_group(model)

    results[model] = {
        'seconds_difference_mean': group_a.seconds_difference.mean(),
        'count': len(group_a.index),
        'age_mean': group_a.age.mean(),
        'age_median': group_a.age.median(),
        'transport': model,
    }

results = pd.DataFrame(results).T

results
# sns.barplot(data=results)
```

```python
sns.barplot(data=results, y='seconds_difference_mean', x='transport', capsize=10)
```

```python
results = {}

for model in grouped.groups:
    for model2 in grouped.groups:
        if model2 == model: continue
        group_a = grouped.get_group(model)
        group_b = grouped.get_group(model2)

        skip = False
        for i in results.keys():
            if model in i.split(':') and model2 in i.split(':'):
                skip = True
                break
        if skip:
            continue
        results[f'{model}:{model2}'] = {}

        f_stat, p_value = stats.f_oneway(group_a.seconds_difference, group_b.seconds_difference)
        results[f'{model}:{model2}']['f-statistic'] = f_stat
        results[f'{model}:{model2}']['p-value'] = p_value

results = pd.DataFrame(results).T
```
